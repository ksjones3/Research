```python
#contains a function to plot stuff
import matplotlib.pyplot as plt

def plot(x, y, xlabel, ylabel, title):
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
#uses pandas to open a file
import pandas as pd

def openCSVFile(inputFilePath):
    data = pd.read_csv(inputFilePath, engine="python", sep="\n", header=None)
    data = data[0].str.split(";", expand=True)
    
    return data
```

```python
class Particle:
    def __init__(self, obj, E, pt, eta, phi):
        self.obj=obj
        self.E=E
        self.pt=pt
        self.eta=eta
        self.phi=phi
    def __init__(self, particleValues):
        self.obj=particleValues[0]
        self.E= particleValues[1]
        self.pt= particleValues[2]
        self.eta= particleValues[3]
        self.phi= particleValues[4]
```

```python
def makeDataList(data):
    dimensions=data.shape
    ymax= dimensions[0]
    xmax= dimensions[1]
    dataList= []
    
    for y in range(0, ymax):
        row = data.iloc[y:y+1]
        eventID = int(row[0])
        processID = row[1]
        eventWeight = float(row[2])
        MET = float(row[3])
        METPhi = float(row[4])
        
        particleList=[]
        
        for x in range(5, xmax):
            particleValues= row[x].str.split(",")[y]
            if particleValues is not None:
                if len(particleValues)==5:
                    obj = particleValues[0]
                    E = particleValues[1]
                    pt = particleValues[2]
                    eta = particleValues[3]
                    phi = particleValues[4]
                
                    particle = Particle(obj, E, pt, eta, phi)
                    particleList.append(particle)
        
        eventDict = {
            "eventID" : eventID,
            "processID" : processID,
            "eventWeight" : eventWeight,
            "MET" : MET,
            "METPhi" : METPhi,
            "particleList" : particleList
        }
        
        dataList.append(eventDict)
    return dataList
```

```python
dataList = makeDataList(data)
```
